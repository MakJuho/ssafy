```
from bs4 import BeautifulSoup
import requests

url = "https://www.consumerreports.org/cro/a-to-z-index/products/index.htm"


# 서버의 response 정보를 이용해 response 객체 생성
response = requests.get(url)

# response 객체에서 html 코드에 해당 부분 추출
data = response.text

# bs4를 이용하여 html 코드를 soup 객체로 파싱
soup = BeautifulSoup(data, 'html.parser')

# 문서 내에 모든 <a>  태그 정보를 찾아 list 로 반환
# tags = soup.find_all('crux-body-copy')
tags =soup.find_all('a')

attr = {'class':'products-a-z__results__item'}
set_dict = soup.find_all('a', attrs=attr)

# print(set_dict)
cnt = 0

for word in set_dict:
    print(soup.select("a.products-a-z__results__item")[cnt].text)
    print('https://www.consumerreports.org/'+word.get('href'))
    cnt+=1

```
