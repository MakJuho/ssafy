### Q1. 코딩뱃('http://codingbat.com/java') 사이트에서 java 카테고리의 문제 묶음 url 링크 정보 조회
* UserAgent 라이브러리를 이용해 chrome 에서 접속한 것처럼 header 정보 생성

```
# 'http://codingbat.com/java' 에 조회되는 html 정보를 이용해 soup 객체

# 조회되는 html 정보 중 문제 묶음 url 링크 정보가 담겨있는 태그 객체 조회
url = "http://codingbat.com/java"
# 서버의 response 정보를 이용해 response 객체 생성
response = requests.get(url)

# response 객체에서 html 코드에 해당 부분 추출
data = response.text

# bs4를 이용하여 html 코드를 soup 객체로 파싱
soup = BeautifulSoup(data, 'html.parser')

# 문서 내에 모든 <a>  태그 정보를 찾아 list 로 반환
attr = {'class':'summ'}
set_dict = soup.find_all('div',attrs=attr)
cnt = 0
all_links = []
for word in set_dict:
    print('https://codingbat.com'+word.select('a')[0]['href'],end="\n\n")
    all_links.append('https://codingbat.com'+word.select('a')[0]['href'])
```
